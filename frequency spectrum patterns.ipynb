```python
import itertools
import numpy as np

# Generate all possible combinations of v with length 9 containing 0s and 1s
v_combinations = list(itertools.product([0, 1], repeat=9))

# Function to calculate the frequency spectrum using Fourier transform
def calculate_frequency_spectrum(v):
    n = len(v)
    fft_result = np.fft.fft(v)
    magnitude_spectrum = np.abs(fft_result) / n
    return magnitude_spectrum

# Calculate frequency spectra for all combinations of v
frequency_spectra = [calculate_frequency_spectrum(v) for v in v_combinations]

# Find combinations of v with the same frequency spectrum
same_spectrum_combinations = {}
for i, spectrum in enumerate(frequency_spectra):
    spectrum_key = tuple(spectrum)  # Convert the spectrum to a tuple to use it as a dictionary key
    if spectrum_key in same_spectrum_combinations:
        same_spectrum_combinations[spectrum_key].append(i)
    else:
        same_spectrum_combinations[spectrum_key] = [i]

# Print combinations of v with the same frequency spectrum
print("Combinations with the same frequency spectrum:")
for combinations in same_spectrum_combinations.values():
    if len(combinations) > 1:
        for index in combinations:
            v = v_combinations[index]
            print(f"Index: {index}, v: {v}, Spectrum: {frequency_spectra[index]}")
        print("---")
```
